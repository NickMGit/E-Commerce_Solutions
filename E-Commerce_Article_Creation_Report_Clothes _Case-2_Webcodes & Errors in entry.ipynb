#### Import statements

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:82.0) Gecko/20100101 Firefox/82.0' } 
```

#### Page with all articles listed

```python
url = "https://www.armani.com/it-it/armani-exchange/donna/abbigliamento/?_gl=1*1f54wr3*_up*MQ..&department=EU_AX_W_Clothing&departmentId=3074457345616697339&gclid=CjwKCAjwuqiiBhBtEiwATgvixAes6bBZ_iSQ8B5adFhWqskl6YC49YlNMbBcNYk7FzZfBLJt3Jc3yxoCDxAQAvD_BwE&gclsrc=aw.ds&itemsToLoadOnNextPage=36&lazyLoadStart=4&linkdepartment=EU_AX_W_Clothing&linkdepartmentId=3074457345616697339&page=3&partialLoadedItems=256&productsPerPage=700&rsiUsed=false&suggestion=false&totalItems=767&totalPages=3&ytosQuery=true"
```

#Parse through html code and place in a list both the article code and its omologue web identifier

```python
products_data_list = []
page = requests.get(url,headers = headers)
soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all("item-card", class_="item-card is-multicolors")
results_1 = soup.find_all("item-card", class_="item-card")

for x in range(len(results_1)):
    products_data_list.append(results_1[x].get("data-ytos-track-product-data"))
for x in range(len(results)):
    products_data_list.append(results[x].get("data-ytos-track-product-data"))

len(products_data_list)
```

#### Create a list of dicts by with both adjacent codes

```python
list_dicts = []
for x in products_data_list:
    list_dicts.append(json.loads(x))
```

#### Separate the codes in adjacent lists

```python
list_values = []

for y in list_dicts:
    list_values.append({y["product_id"]:y["product_mfPartNumber"]})

cod_foto = []
cod_articolo = []
    
for dct in list_values:
    cod_foto.append(list(dct.keys()))
for vlu in list_values:
    cod_articolo.append(list(vlu.values()))
```

#### Join nested lists

```python
cod_foto_list = [''.join(ele) for ele in cod_foto]
cod_articolo_list = [''.join(eles) for eles in cod_articolo]
```

#### Create Dataframe

```python
df_linked_codes = pd.DataFrame()
```

#### Transform lists into dataframe columns

```python
df_linked_codes["Cod_foto_url"] = cod_foto_list
df_linked_codes["ModelloVariante"] = cod_articolo_list
```

#### Import the stock Dataframe

```python
df_atelier = pd.read_excel("Armani Exchange Stock 27.04.2023.xls")
```

#### Set Titles as column headers, get rid of meta data

```python
df_atelier.columns = df_atelier.iloc[3]
df_atelier = df_atelier[4:]
```

#### Drop specific row

##### Get column from stock data codes as a list

```python
atelier_modello = list(df_atelier["ModelloVariante"])
```

#### Create dataframe for code comparisons

```python
df_look_alikes = pd.DataFrame()
```

#### Make lists out of Article codes from stock data, article codes from the web, resemblence percentage

```python
code_atelier_list = []
code_website_list = []
resemblence_list = []

for xkl in range(len(df_linked_codes["ModelloVariante"])):
        for codice in df_linked_codes["ModelloVariante"]:
            if fuzz.ratio(atelier_modello[xkl],codice) > 95: 
                code_atelier_list.append(str(atelier_modello[xkl]))
                code_website_list.append(str(codice))
                resemblence_list.append(str(fuzz.ratio(atelier_modello[xkl],codice)))
```

#### Make a datafraem from the previous lists

```python
df_look_alikes["Code Atelier"]= code_atelier_list
df_look_alikes["ModelloVariante"]= code_website_list
df_look_alikes["Resemblence %"]= resemblence_list    
df_look_alikes                                        
```

#### export linked codes as .CSV 

```python
df_linked_codes.to_csv("df_linked")
```

#### Concatenate article codes/resemblence & linked article codes/web codes dataframes on the commom ones

```python
concat_df = pd.merge(df_linked_codes,df_look_alikes,on="ModelloVariante",how="inner")
```

#### Rename column from stock data to merge with previous concatenated dataframe

```python
df_atelier = df_atelier.rename(columns={"ModelloVariante":"Code Atelier"})
```
